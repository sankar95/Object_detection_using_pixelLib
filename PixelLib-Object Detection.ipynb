```python
import pixellib
from pixellib.instance import instance_segmentation
import cv2
```

# 2. Setup Model

```python
seg_model = instance_segmentation()
seg_model.load_model("mask_rcnn_coco.h5")
```

# 3. Real Time Capture

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame = cap.read()
    
    # Apply instance segmentation
    res = seg_model.segmentFrame(frame,show_bboxes=True)
    image=res[1]
    cv2.imshow("Sai baba", image)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
